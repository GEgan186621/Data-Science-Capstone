```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scores = {
    "Model": ["Decision Tree", "KNN", "Logistic Regression", "SVM"],
    "CV Score": [0.875, .8482142857142858, 0.8464285714285713, 0.8482142857142856]   # typed in values
}
df = pd.DataFrame(scores)
print(df)
```

```python

df = df.sort_values(by="CV Score", ascending=False)

plt.figure()
plt.bar(df["Model"], df["CV Score"])
plt.xticks(rotation=30)
plt.title("Model Performance Comparison")
plt.show()
```

```python
plt.figure()
plt.bar(df["Model"], df["CV Score"])

plt.ylim(0.84, 0.88)  # zoom

plt.ylabel("Cross-Validation Score")
plt.title("Model Performance (Zoomed View)")
plt.xticks(rotation=30)

plt.show()
```

```python

import matplotlib.pyplot as plt

plt.figure()
plt.bar(df["Model"], df["CV Score"])
plt.xlabel("Model")
plt.ylabel("Cross-Validation Score")
plt.title("Model Comparison")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

df = df.sort_values(by="CV Score", ascending=False)
```
